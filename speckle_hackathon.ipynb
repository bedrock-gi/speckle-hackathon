```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

voxel_edge_length_meters = 5
number_of_paths_to_generate = 10
step_length_meters = 100
triangle_floor_half_length_meters = 20
triangle_floor_interval_length_meters = 5
assert triangle_floor_half_length_meters % triangle_floor_interval_length_meters == 0
maximum_number_of_descendants_per_node = 2 * int(triangle_floor_half_length_meters / triangle_floor_interval_length_meters) + 1

a_x_meters = 
a_y_meters = 
b_x_meters = 
b_y_meters = 

class Node(object):
    def __init__(self, x_meters, y_meters, parent_node=None, end_node=False):
        self.x_meters = x_meters
        self.y_meters = y_meters
        self.parent_node = parent_node
        self.end_node = end_node
        if self.parent_node is not None:
            self.slope_of_the_line_segment = (self.y_meters - self.parent_node.y_meters) / (self.x_meters - self.parent_node.x_meters)
            self.intercept_of_the_line_segment = self.y_meters - self.slope_of_the_line_segment * self.x_meters
        self.cost_of_only_the_current_node = self.get_cost_of_only_the_current_node()
        self.cost_upto_the_current_node = self.get_cost_upto_the_current_node()
        direction_to_b_x_meters = b_x_meters - x_meters
        direction_to_b_y_meters = b_y_meters - y_meters
        self.distance_to_b_meters = np.sqrt(np.power(direction_to_b_x_meters, 2) + np.power(direction_to_b_y_meters, 2))
        if self.distance_to_b_meters > step_length_meters:
            unit_direction_to_b_x = direction_to_b_x_meters / self.distance_to_b_meters
            unit_direction_to_b_y = direction_to_b_y_meters / self.distance_to_b_meters
            self.scaled_direction_to_b_x_meters = step_length_meters * unit_direction_to_b_x
            self.scaled_direction_to_b_y_meters = step_length_meters * unit_direction_to_b_y
            
            orthogonal_to_unit_direction_to_b_x_meters = 1.0
            orthogonal_to_unit_direction_to_b_y_meters = -unit_direction_to_b_x / unit_direction_to_b_y
            length_of_orthogonal_to_unit_direction_to_b_meters = np.sqrt(1.0 + np.power(-unit_direction_to_b_x / unit_direction_to_b_y, 2))
            self.unit_orthogonal_to_unit_direction_to_b_x = orthogonal_to_unit_direction_to_b_x_meters / length_of_orthogonal_to_unit_direction_to_b_meters
            self.unit_orthogonal_to_unit_direction_to_b_y = orthogonal_to_unit_direction_to_b_y_meters / length_of_orthogonal_to_unit_direction_to_b_meters
            self.scaled_orthogonal_to_unit_direction_to_b_x_meters = triangle_floor_interval_length_meters * self.unit_orthogonal_to_unit_direction_to_b_x
            self.scaled_orthogonal_to_unit_direction_to_b_y_meters = triangle_floor_interval_length_meters * self.unit_orthogonal_to_unit_direction_to_b_y

        self.descendant_id_descendant_node_mapping = dict()

    def get_cost_upto_the_current_node(self):
        current_node = self
        current_cost = current_node.cost_of_only_the_current_node
        while current_node.parent_node is not None:
            current_node = current_node.parent_node
            current_cost *= current_node.cost_of_only_the_current_node
        return current_cost

    def get_cost_of_only_the_current_node(self):
        if self.parent_node is None:
            cost_of_only_the_current_node = 1.0
        else:
            cost_of_only_the_current_node = 1.0

            current_point = self.parent_node
            current_point_x = self.parent_node.x_meters
    
            final_point = self
            final_point_x = final_point.x_meters
    
            if current_point_x < final_point_x: # increase current_point_x until it is larger than final_point_x
                while current_point_x < final_point_x:
                    current_point_y = self.slope_of_the_line_segment * current_point_x + self.intercept_of_the_line_segment
                    current_point_voxel_x_coordinate = int(current_point_x // voxel_edge_length_meters)
                    current_point_voxel_y_coordinate = int(current_point_y // voxel_edge_length_meters)
                    cost_of_only_the_current_node *= cost_array[current_point_voxel_y_coordinate][current_point_voxel_x_coordinate]
                    current_point_x += voxel_edge_length_meters
            else: # decrease current_point_x until it is lower than final_point_x
                while current_point_x > final_point_x:
                    current_point_y = self.slope_of_the_line_segment * current_point_x + self.intercept_of_the_line_segment
                    current_point_voxel_x_coordinate = int(current_point_x // voxel_edge_length_meters)
                    current_point_voxel_y_coordinate = int(current_point_y // voxel_edge_length_meters)
                    cost_of_only_the_current_node *= cost_array[current_point_voxel_y_coordinate][current_point_voxel_x_coordinate]
                    current_point_x -= voxel_edge_length_meters
        return cost_of_only_the_current_node

    def generate_new_descendant(self):
        if self.distance_to_b_meters <= step_length_meters:
            if len(self.descendant_id_descendant_node_mapping.keys()) == 0:
                new_descendant_id = 0
                new_descendant = Node(x_meters=b_x_meters, y_meters=b_y_meters, parent_node=self, end_node=True)
                self.descendant_id_descendant_node_mapping[new_descendant_id] = new_descendant
                return new_descendant
            else:
                return None
        else:
            if len(self.descendant_id_descendant_node_mapping.keys()) == 0:
                new_descendant_id = 0
            else:
                new_descendant_id = max(list(self.descendant_id_descendant_node_mapping.keys())) + 1
    
            if new_descendant_id == maximum_number_of_descendants_per_node:
                return None
            else:
                if new_descendant_id == 0:
                    new_descendant = Node(x_meters=self.x_meters + self.scaled_direction_to_b_x_meters, y_meters=self.y_meters + self.scaled_direction_to_b_y_meters, parent_node=self, end_node=False)
                elif new_descendant_id % 2 == 1:
                    new_descendant = Node(x_meters=self.descendant_id_descendant_node_mapping[0].x_meters + ((new_descendant_id + 1) // 2) * self.scaled_orthogonal_to_unit_direction_to_b_x_meters, y_meters=self.descendant_id_descendant_node_mapping[0].y_meters + self.scaled_orthogonal_to_unit_direction_to_b_y_meters, parent_node=self, end_node=False)
                elif new_descendant_id % 2 == 0:
                    new_descendant = Node(x_meters=self.descendant_id_descendant_node_mapping[0].x_meters - ((new_descendant_id + 1) // 2) * self.scaled_orthogonal_to_unit_direction_to_b_x_meters, y_meters=self.descendant_id_descendant_node_mapping[0].y_meters - self.scaled_orthogonal_to_unit_direction_to_b_y_meters, parent_node=self, end_node=False)

                self.descendant_id_descendant_node_mapping[new_descendant_id] = new_descendant
                return new_descendant


start_node = Node(x_meters=a_x_meters, y_meters=a_y_meters, parent_node=None, end_node=False)

nodes_to_explore = [start_node]
current_final_nodes = []
while len(current_final_nodes) < number_of_paths_to_generate:
    node_to_explore = nodes_to_explore[-1]
    new_descendant = node_to_explore.generate_new_descendant()
    if new_descendant is None:
        nodes_to_explore.pop()
    elif new_descendant.end_node:
        current_final_nodes.append(new_descendant)
    elif new_descendant.cost_upto_the_current_node == 0.0:
        continue
    else:
        nodes_to_explore.append(new_descendant)

# Find the final node with minimum cost
min_cost = np.inf
final_node_with_min_cost = None
for current_final_node in current_final_nodes:
    cost_of_current_final_node = current_final_node.cost_upto_the_current_node
    if cost_of_current_final_node < min_cost:
        min_cost = cost_of_current_final_node
        final_node_with_min_cost = current_final_node

# Plot paths, right bottom: start of the tunnel, left top: end of the tunnel
for current_node in current_final_nodes:
    plt.figure(figsize=(18, 6))
    while current_node.parent_node is not None:
        plt.plot([current_node.x_meters, current_node.parent_node.x_meters], [current_node.y_meters, current_node.parent_node.y_meters], "go-")
        current_node = current_node.parent_node

# Print the selected path from start point to end point
current_node = final_node_with_min_cost
path = [(current_node.x_meters, current_node.y_meters)]
while current_node.parent_node is not None:
    current_node = current_node.parent_node
    path = [(current_node.x_meters, current_node.y_meters)] + path
    print(path)
```
